```
train_data = pandas.read_csv(content_path+'train_data(fixed).csv')
```

```
train_data = train_data.drop(columns=['Column1'])
```

```
print(train_data)
```

```
print(train_data.groupby('Label').size())
```

```
stop_word = ['+','-','=','_','{','}','[',']',':',';','"',"'",'<','>',',','.','?','/','|','\\','!','@','#','$','%','^','&','*','(',')',
             'bị','bởi','cả','các','càng','cần','và','rằng','chứ','là','lại','một_cách','bị','gì','mà','cũng','bia','biển','100k','10K','15','10','ai','anh','ba','bà']
```

```
tmp = train_data['Data'].to_list()
```

```
print(len(tmp))
```

```
for idx, text in enumerate(tmp):
  for sw in stop_word:
    if sw in text:
      tmp[idx] = tmp[idx].replace(sw,' ')
```

```
print(tmp[0])
```

```
vectors = TfidfVectorizer(max_df=0.8,min_df=0.01)
```

```
vectors.fit_transform(tmp)
```

```
vectors
```

```
train_label = train_data['Label']
```

```
from sklearn.naive_bayes import MultinomialNB
```

```
MBN_model = MultinomialNB()
```

```
MBN_model.fit(vectors.idf_, train_label)
```
